```python
datac['OverAll_PassStatus'] = datac.apply(lambda x : 'F' if x['Math_PassStatus'] == 'F' or 
                                    x['Reading_PassStatus'] == 'F' or x['Writing_PassStatus'] == 'F' else 'P', axis =1)

datac.OverAll_PassStatus.value_counts()
```

```python
p = sns.countplot(x='parental level of education', data = datac, hue='OverAll_PassStatus', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=20) 
```

```python
datac['Total_Marks'] = datac['math score']+datac['reading score']+datac['writing score']
datac['Percentage'] = datac['Total_Marks']/3
```

```python
plt.figure(figsize=(16,6))
p = sns.countplot(x="Percentage", data = datac, palette="muted")
_ = plt.setp(p.get_xticklabels(), rotation=0) 
plt.tight_layout()
plt.show()
```

```python
def GetGrade(Percentage, OverAll_PassStatus):
    if ( OverAll_PassStatus == 'F'):
        return 'F'    
    if ( Percentage >= 80 ):
        return 'A'
    if ( Percentage >= 70):
        return 'B'
    if ( Percentage >= 60):
        return 'C'
    if ( Percentage >= 50):
        return 'D'
    if ( Percentage >= 40):
        return 'E'
    else: 
        return 'F'

datac['Grade'] = datac.apply(lambda x : GetGrade(x['Percentage'], x['OverAll_PassStatus']), axis=1)

datac.Grade.value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.set_context("talk",font_scale=1)
sns.set_palette("pastel")
ax = sns.countplot(y="Grade", hue="gender", data=datac, order=["A","B","C","D","E","F"])
ax.legend(loc='upper right',frameon=True)
plt.title('Gender vs Grades', fontsize=18, fontweight='bold')
ax.set(xlabel='COUNT',ylabel='GRADE')
plt.show()
```
